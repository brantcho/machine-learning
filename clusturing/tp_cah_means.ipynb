```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fromage = pd.read_table("fromage.txt", sep="\t", header=0, index_col=0)
fromage.head()
```

```python
fromage.shape
```

```python
fromage.describe()
```

```python
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
# Centrer et réduire
fromage_cr = StandardScaler(with_mean=False, with_std=True).fit_transform(fromage)
#fromage_cr = pd.DataFrame(data=fromage_cr, columns = fromage.columns, index = fromage.index)
# Calculer les dissimilarités
dist = linkage(fromage_cr, method='ward', metric='euclidean')
# Faire le dendogramme
plt.figure()
dendrogram(dist, labels=fromage.index, orientation='left', color_threshold=7)

plt.show()
```
