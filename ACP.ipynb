```python
# matrice des variances des differentes colonnes 
S= Y.var(axis = 0)
S
```

```python
n = X.shape[0]
D = np.diag(np.ones(n)) 
D = D/n
D
```

```python
M = np.diag(1/np.sqrt(S))
M
```

```python
V=Y.T.dot(D).dot(Y)
V
```

```python
VM = V.dot(M)
VM
```
